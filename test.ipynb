```python
df_filter = df[~df['TRAVEL_MISSION_CHECK'].isna()].copy()

df_filter.loc[:, 'TRAVEL_MISSION_INT'] = df_filter['TRAVEL_MISSION_CHECK'].str.split(';').str[0].astype(int)
```

```python
df_filter = df_filter[[
    'GENDER',
    'AGE_GRP',
    'TRAVEL_STYL_1',
    'TRAVEL_STYL_2',
    'TRAVEL_STYL_3',
    'TRAVEL_STYL_4',
    'TRAVEL_STYL_5',
    'TRAVEL_STYL_6',
    'TRAVEL_STYL_7',
    'TRAVEL_STYL_8',
    'TRAVEL_MOTIVE_1',
    'TRAVEL_COMPANIONS_NUM',
    'TRAVEL_MISSION_INT',
    'VISIT_AREA_NM',
    'DGSTFN'
]]

df_filter = df_filter.dropna()

df_filter.info()
```

```python
df_filter.head()
```

```python
categorical_features_names = [
    'GENDER',
    # 'AGE_GRP',
    'TRAVEL_STYL_1',
    'TRAVEL_STYL_2',
    'TRAVEL_STYL_3',
    'TRAVEL_STYL_4',
    'TRAVEL_STYL_5',
    'TRAVEL_STYL_6',
    'TRAVEL_STYL_7',
    'TRAVEL_STYL_8',
    'TRAVEL_MOTIVE_1',
    # 'TRAVEL_COMPANIONS_NUM',
    'TRAVEL_MISSION_INT',
    'VISIT_AREA_NM',
    # 'DGSTFN'
]

df_filter[categorical_features_names[1:-1]] = df_filter[categorical_features_names[1:-1]].astype(int)

df_filter.head()


```

```python
df_filter.info()
```

```python
from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(df_filter, test_size=0.2, random_state=42)
print(train_df.shape)
print(test_df.shape)
```

```python
from catboost import CatBoostRegressor, Pool

train_pool = Pool(
    train_df.drop(['DGSTFN'], axis=1),
    label=train_df['DGSTFN'],
    cat_features=categorical_features_names
)

test_pool = Pool(
    test_df.drop(['DGSTFN'], axis=1),
    label=test_df['DGSTFN'],
    cat_features=categorical_features_names
)
```

```python
model = CatBoostRegressor(
    loss_function='RMSE',
    eval_metric='MAE',
    task_type='CPU',
    depth=6,
    learning_rate=0.01,
    n_estimators=2000,
)

model.fit(
    train_pool,
    eval_set=test_pool,
    verbose=500,
    plot=True
)
```
